```python
#multiplication table of any number
n = int(input("enter num: "))
i = 1
while(i<=10):
        print(n*i)
        i += 1
```

```python
#break 
i=1
while(i<=10):
    if(i%6==0):
      break
    print(i)
    i +=1
    print("outside loop now....")
```
