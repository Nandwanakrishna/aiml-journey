```python
#Q1 Ask the user for a string and check whether it is a palindrome or not A  palindrome is a string which is same when we read it forward & backward. Eg -"madam","racecar" etc. 
text = input("Enter a string: ")
reverse = text[::-1]
if text == reverse:
    print("palidrome")
else:
    print("not palidrome")
```

```python
#Q6Given a list of words:
#words = ["apple", "banana", "kiwi", "cherry", "mango"]
#Create a dictionary that maps each word to its length.
#Example:{"apple": 5, "banana": 6, "kiwi": 4, ...


words = ["apple", "banana", "kiwi", "cherry", "mango"]
word_length = {}
for word in words:
    word_length[words] = len(word)
    print(word_length)
```

```python
#Q9Given a list, print all elements that appear more than once in the list.
numbers = [1,2,3,2,4,3,5]
duplicates = []
for n in numbers:
    if n.count(n)>1:
        if n not in duplicates:
            duplicates .append(n)
            print("Duplicates element: ",duplicates)
```
